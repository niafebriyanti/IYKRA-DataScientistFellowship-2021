##Data Preparation

```
#Import library
import pandas as pd
import numpy as np
```

```
#Read the dataset
data = pd.read_csv("https://raw.githubusercontent.com/Syukrondzeko/Fellowship-5/main/Seasons_Stats.csv")
display(data.shape)
data.head(20)
```

```
#Brief information of the dataset
data.info()
```

```
#Brief statistics
data.describe()
```

```
#Show percentage of null values 
data.isnull().sum()*100/data.shape[0]
```

```
#Delete columns that have as many missing values as the entire row of data
data_clean = data.drop(columns = ['blanl','blank2'])
```

```
#Show the null values
data_clean.isnull().sum()*100/data_clean.shape[0]
```

```
#Change data type
data_clean['Year'] = data_clean['Year'].astype("Int64")
data_clean['Age'] = data_clean['Age'].astype("Int64")

#Take only 2017 data
data_2017 = data_clean[data_clean['Year']==2017]
```

```
#Cleaning duplicate value
data_2017 = data_2017.drop_duplicates(subset = ['Player', 'Age'], keep='last')
data_2017.head()
```

##Questions

> Question 1.

- Who is the youngest and oldest player in the NBA in 2017 for each team (Tm) ?

######The youngest player each team

```
data_2017['Tm'].unique()
```

```
#create empty dataframe
minimum = pd.DataFrame(columns=['Tm','Player','Age'])

team = data_2017['Tm'].unique() ##get the unique team in 2017
Q1 = data_2017[['Tm','Player','Age']] ##get only Team, player and age variable

for i in team:
  team_1 = Q1[Q1['Tm'] == i]
  team_2 = team_1[team_1['Age'] == team_1['Age'].min()]
  minimum = minimum.append(team_2, ignore_index=False)

#show the result
minimum.reset_index(drop=True)
```

```
#create empty dataframe
maximum = pd.DataFrame(columns=['Tm','Player','Age'])

team = data_2017['Tm'].unique() ##get the unique team in 2017
Q1 = data_2017[['Tm','Player','Age']] ##get only Team, player and age variable

for i in team:
  team_1 = Q1[Q1['Tm'] == i]
  team_2 = team_1[team_1['Age'] == team_1['Age'].max()]
  maximum = maximum.append(team_2, ignore_index=False)

maximum.reset_index(drop=True)
```

> Question 2.

- Which player has the most minutes played (MP) in each position (Pos)?

```
Q2 = data_2017.copy()

#take Player, Position, MP Variable
Q2 = Q2[['Player',"Pos","MP"]]

#take unique Position
position = Q2['Pos'].unique()

#create empty dataframe
most_MP = pd.DataFrame(columns=['Player',"Pos","MP"])

#Display all player in each position who has the most Minutes Played (MP)
for i in position:
  Q2_position = Q2[Q2['Pos'] == i]
  Q2_max_mp = Q2_position[Q2_position['MP'] == Q2_position['MP'].max()]
  most_MP = most_MP.append(Q2_max_mp, ignore_index=False)

most_MP.reset_index(drop=True)
```

> Question 4.

- Who is the best player in your opinion based on his record stats? note: you can refer to variables point (PTS), assists, rebounds, or anything else. A combination of several variables would be nice.

###Basketball Statistics Definition
Can be found [here](https://www.basketball-reference.com/about/glossary.html)

GM, GP; GS: games played; games started

PTS: points

FGM, FGA, FG%: field goals made, attempted and percentage

FTM, FTA, FT%: free throws made, attempted and percentage

3FGM, 3FGA, 3FG%: three-point field goals made, attempted and percentage

REB, OREB, DREB: rebounds, offensive rebounds, defensive rebounds

AST: assists

STL: steals

BLK: blocks

TO: turnovers

EFF: efficiency: NBA's efficiency rating: (PTS + REB + AST + STL + BLK − ((FGA − FGM) + (FTA − FTM) + TO))

PF: personal fouls

MIN: minutes

AST/TO: assist to turnover ratio

PER: Player Efficiency Rating: John Hollinger's Player Efficiency Rating

PIR: Performance Index Rating: Euroleague's and Eurocup's Performance Index Rating.

PIR: (Points + Rebounds + Assists + Steals + Blocks + Fouls Drawn) − (Missed Field Goals + Missed Free Throws + Turnovers + Shots Rejected + Fouls Committed)

A journal about PIR formula can be found [here](http://jdatasci.com/index.php/jdatasci/article/download/3/4/)

```
data_2017.columns
```

```
Q4 = data_2017.copy()
Q4['A'] = Q4['PTS'] + Q4['TRB'] + Q4['AST'] + Q4['STL'] + Q4['BLK'] + Q4['PF']
Q4['B'] = Q4['FG'] + Q4['FT'] + Q4['TOV']
Q4['PIR_Value'] = Q4['A'] - Q4['B']

best_player = Q4[['Player','Tm','PIR_Value']]
best_player[best_player['PIR_Value'] == best_player['PIR_Value'].max()]
```

> Question 5

- Which team has the best average stat record of their players? Note: you can refer to points, assists, rebounds, or anything else. A combination of several variables would be nice.

To find the best team, I use the PIR value from each player each team and then average the PIR for each player from each team.

```
#The new variable to know the best stat is PIR_Value
Q5 = Q4.copy()

#Grouping by team
Q5 = Q5[['Tm','PIR_Value']]
best_team = Q5.groupby(['Tm']).mean().reset_index()

#show the best team 
best_team[best_team['PIR_Value'] == best_team['PIR_Value'].max()]
#best_team.sort_values('PIR_Value', ascending=False)
```
